```python

a = int(input('Enter the first number: '))
b = int(input('Enter the second number: '))
c = int(input('Enter the third number: '))

# calculate discriminant
discriminant = (b**2 - 4*a*c)

# check for real roots
if discriminant >= 0:
    root1 = (-b + (discriminant)**0.5) / (2*a)
    root2 = (-b - (discriminant)**0.5) / (2*a)
    print("Roots are:", int(root1), "and", int(root2))
else:
    print("No real roots")
```

```python
from pylatex import Document, Section, Math, Matrix
from pylatex.utils import italic

# Create document with geometry package for margins
doc = Document(geometry_options={"tmargin": "1cm", "lmargin": "2cm"})

# Add title section
with doc.create(Section("Essential Machine Learning Formulas")):
    doc.append("This document contains fundamental mathematical formulas used in machine learning and deep learning.")

# Distance Metrics section
with doc.create(Section("Distance Metrics")):
    # Euclidean Distance
    doc.append("Euclidean distance between two vectors:")
    doc.append(Math(data=[r"d(\mathbf{x}, \mathbf{y}) = \sqrt{\sum_{i=1}^n (x_i - y_i)^2}"]))
    
    # Manhattan Distance
    doc.append("Manhattan distance formula:")
    doc.append(Math(data=[r"d(\mathbf{x}, \mathbf{y}) = \sum_{i=1}^n |x_i - y_i|"]))
    
    # Cosine Similarity
    doc.append("Cosine similarity between vectors:")
    doc.append(Math(data=[r"\text{similarity}(\mathbf{A},\mathbf{B}) = \frac{\mathbf{A} \cdot \mathbf{B}}{\|\mathbf{A}\| \|\mathbf{B}\|}"]))

# Activation Functions section
with doc.create(Section("Activation Functions")):
    # ReLU
    doc.append("Rectified Linear Unit (ReLU):")
    doc.append(Math(data=[r"f(x) = \max(0, x)"]))
    
    # Sigmoid
    doc.append("Sigmoid activation function:")
    doc.append(Math(data=[r"\sigma(x) = \frac{1}{1 + e^{-x}}"]))
    
    # Softmax
    doc.append("Softmax function for probability distribution:")
    doc.append(Math(data=[r"\sigma(\mathbf{z})_i = \frac{e^{z_i}}{\sum_{j=1}^K e^{z_j}}"]))

# Loss Functions section
with doc.create(Section("Loss Functions")):
    # MSE
    doc.append("Mean Squared Error:")
    doc.append(Math(data=[r"\text{MSE} = \frac{1}{n}\sum_{i=1}^n (y_i - \hat{y}_i)^2"]))
    
    # Cross-Entropy
    doc.append("Cross-Entropy Loss:")
    doc.append(Math(data=[r"L = -\sum_{c=1}^C y_c \log(\hat{y}_c)"]))
    
    # Hinge Loss
    doc.append("Hinge Loss for SVMs:")
    doc.append(Math(data=[r"\ell(y) = \max(0, 1 - t \cdot y)"]))

# Generate PDF
doc.generate_pdf("ml_formulas", clean_tex=False)
```
